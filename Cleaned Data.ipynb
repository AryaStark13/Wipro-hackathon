# Importing and cleaning the Train set
Performing this on standardized data

```python
train_set = pd.read_csv(r"C:\Users\ariha\Desktop\Hackathons & Projects\Wipro Hackathon\train.csv")
train_set
```

```python
test_set = pd.read_csv(r"C:\Users\ariha\Desktop\Hackathons & Projects\Wipro Hackathon\test.csv")
test_set
```

```python
y_train = pd.DataFrame(train_set[['Clearsky DHI', 'Clearsky DNI', 'Clearsky GHI']])
y_train
```

```python
y_test = pd.DataFrame(test_set[['Clearsky DHI', 'Clearsky DNI', 'Clearsky GHI']])
y_test
```

```python
x_train = train_set.drop(columns=['Clearsky DHI', 'Clearsky DNI', 'Clearsky GHI'], axis=1)
x_train
```

```python
x_test = test_set.drop(columns=['Clearsky DHI', 'Clearsky DNI', 'Clearsky GHI'], axis=1)
x_test
```

```python
x_train['Date']=pd.to_datetime(x_train[['Year', 'Month', 'Day', 'Hour', 'Minute']])
x_train
```

```python
x_train.drop(columns=['Year', 'Month', 'Day', 'Hour', 'Minute'], axis=1, inplace=True)
x_train
```

```python
x_test['Date']=pd.to_datetime(x_test[['Year', 'Month', 'Day', 'Hour', 'Minute']])
x_test.drop(columns=['Year', 'Month', 'Day', 'Hour', 'Minute'], axis=1, inplace=True)
x_test
```

```python
corr_matrix = x_train.corr()
corr_matrix
```

```python
import numpy as np
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(np.bool))
upper
```

```python
columns_to_drop = []
for cols in upper.columns:
    if any (upper[cols] > 0.75):
        columns_to_drop.append(cols)
columns_to_drop
```

```python
x_train.drop(columns=columns_to_drop, axis=1, inplace=True)
x_train.head()
```

```python
x_test.drop(columns=columns_to_drop, axis=1, inplace=True)
x_test.head()
```

```python
columns_list = []
for cols in x_train.columns:
    columns_list.append(cols)
    
columns_list
```

```python
columns_list = []
for cols in x_test.columns:
    columns_list.append(cols)
    
columns_list
```

```python
columns_list.remove('Date')
columns_list.insert(0, 'Date')
columns_list
```

```python
columns_list.remove('Date')
columns_list.insert(0, 'Date')
columns_list
```

```python
x_train = x_train[columns_list]
x_train.head()
```

```python
x_test = x_test[columns_list]
x_test.head()
```

```python
x_train['Fill Flag'].value_counts()
```

```python
for cols in x_train:
    if cols != 'Date':
        x_train[cols] = (x_train[cols] - x_train[cols].mean()) / x_train[cols].std()
x_train
```

```python
for cols in x_train:
    if cols != 'Date':
        x_test[cols] = (x_test[cols] - x_test[cols].mean()) / x_test[cols].std()
x_test
```

```python
for cols in y_train:
    y_train[cols] = (y_train[cols] - y_train[cols].mean()) / y_train[cols].std()
y_train
```
